## 作業目標


* 根據範例使用 API 取出 高雄市小港區 2019 年 空氣品質 資料，並用一個適合的資料結構儲存
* Hint: 將所有資料存在一個變數當中


```python
import requests
import json
url = 'https://airtw.epa.gov.tw/ajax.aspx'
payload = {
    'Target': 'air_list',
    'SiteID': '9',
    'Datatime': '2020/04/12 19:00:00',
    'Type': 'home'
}
r = requests.post(url, data=payload)
#r.status_code
r.text
```

新的網站用日期來要資料


```python

url = 'https://airtw.epa.gov.tw/ajax.aspx'
for i in range(12):    #12
    payload = {
        'Target': 'air_list',
        'SiteID': '9',
        'Datatime': '2020/04/'+ str(12 - i)+' 19:00:00',
        'Type': 'home'
    }
    r = requests.post(url, data=payload)
    j = json.loads(r.text)
    print(j)
```
