## 作業目標

利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。


### 1. https://www.zhihu.com/explore 

```python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

### 利用查看開發者工具觀察 Headers



```python
import requests
url = 'http://www.zhihu.com'

headers = {
    "access-control-allow-origin:"" *"
    "cache-control: max-age=31536000"
    "content-encoding: gzip"
    "content-length: 2929"
    "content-type: text/css"
    "date: Tue, 07 Apr 2020 13:26:01 GMT"
    "expires: Wed, 07 Apr 2021 13:26:01 GMT"
    "last-modified: Tue, 07 Apr 2020 07:51:30 GMT"
    "server: NWS_Oversea_AP"
    "status: 200"
    "timing-allow-origin: *"
    "x-cache-lookup: Hit From Disktank3 Gz"
    "x-cache-lookup: Hit From Inner Cluster"
    "x-cdn-provider: tencent"
    "x-daa-tunnel: hop_count=1"
    "x-nws-log-uuid: 68a83307-e13a-4e0c-9c7d-385a47c9fbd9"
}

r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:600])
```

### 2. http://aicoin.cn

```python
import requests

url = 'http://aicoin.cn/'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

### 利用查看開發者工具觀察 Headers



```python
import requests
url = 'http://aicoin.cn/'

headers = {
    "Connection": "keep-alive"
    "Content-Encoding:" "gzip"
    "Content-Type:" "text/html; charset=utf-8"
    "Date:" "Tue, 07 Apr 2020 13:40:05 GMT"
    "ETag:" 'W/"16bf15-1RaHg5TQRpf/PQYJPFLLZA6C3TY"'
    "Server:" "CloudWAF"
    "Transfer-Encoding:" "chunked"
}

r = requests.get(url, headers=headers)
print(r.text[0:600])
```
