## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
# 每個兩秒鐘自動往下滑

import time
for i in range(200):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 100000);")
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
from datetime import datetime, timedelta
nowtime = datetime.now()
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    date_strp = datetime.strptime(date, '%Y/%m/%d %H:%M')
    if nowtime - date_strp < timedelta(hours=24):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
# 關閉瀏覽器
browser.quit();
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime, timedelta
nowtime = datetime.now()
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    date_strp = datetime.strptime(date, '%Y/%m/%d %H:%M')
    if nowtime - date_strp < timedelta(hours=2):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime, timedelta
nowtime = datetime.now()
three_oclock=datetime.strptime('2020/04/09 15:00', '%Y/%m/%d %H:%M')
five_oclock=datetime.strptime('2020/04/09 17:00', '%Y/%m/%d %H:%M')
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    date_strp = datetime.strptime(date, '%Y/%m/%d %H:%M')
    if  three_oclock < date_strp < five_oclock:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
