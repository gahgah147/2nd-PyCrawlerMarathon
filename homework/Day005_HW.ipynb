```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
c=0
for d in data:
    c=c+1
    print(c,"標題:",d['title'] ,"\n")
    print("貼文時間:",d['createdAt'] ,"\n")
    print("留言人數:",d['commentCount'] ,"\n")
    print("按讚人數:",d['likeCount'] ,"\n")
```

```python

#先計算總共平均數
commentCount = 0
likeCount =0
for d in data:
    commentCount= commentCount + d['commentCount']
    likeCount =likeCount + d['likeCount']
    
print(commentCount/30 ,",", likeCount/30)
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
# 假設定義 平均按讚人數超過100算熱門文章
          
hot_commentCount = 0
unhot_commentCount = 0

hot_likeCount =0
unhot_likeCount =0

hot_PostCount = 0
unhot_PostCount = 0
for d in data:
    if d['likeCount'] >=100 :
        hot_PostCount = hot_PostCount+1
        hot_commentCount= hot_commentCount + d['commentCount']
        hot_likeCount= hot_likeCount + d['likeCount']
    else:
        unhot_PostCount = unhot_PostCount+1
        unhot_commentCount=unhot_commentCount+d['commentCount']
        unhot_likeCount = unhot_commentCount +d['likeCount']

print("熱門文章平均留言人數:",hot_commentCount/hot_PostCount)
print("熱門文章平均按讚人數:",hot_likeCount/hot_PostCount)
print("非熱門文章平均留言人數:",unhot_commentCount/unhot_PostCount)
print("非熱門文章平均按讚人數:",unhot_likeCount/unhot_PostCount)
```
